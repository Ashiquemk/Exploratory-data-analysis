```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
#we are cobining two tables zomato and country-code
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
#to check data types
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#pie chart of top three countries using pie charts
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

observations:Zomato maximum records or transactions are from India.after that uk and us.


```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# observations
1.when the rating is in the range 4.5 to 4.9 ----> Excellent
2.when the rating is in the range of 4.0 to 4.4 ---->very good

```python
# to get bigger diagrams
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

obsrvations:
    1.Not rated count is very high
    2.maximum number of rating are between 2.5t o 3.4

```python
#find the countries taht has given zero rantings
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

observation
maximum number of 0 rating is from indian customers
